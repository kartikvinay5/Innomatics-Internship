```python
# Task 1 - User Login Check
# Store correct login details
correct_user = "admin"
correct_pass = "1234"
```

```python
entered_user = input("Enter username: ")
entered_pass = input("Enter password: ")
```

```python
# Check if login details match
if entered_user == correct_user and entered_pass == correct_pass:
    print("Login Successful")
else:
    print("Invalid Credentials")

```

```python
# Task 2 - Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
```

```python
pass_count = 0
fail_count = 0
```

```python
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1
```

```python
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
# Task 3 - Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]
```

```python
cleaned_names = []
```

```python
for name in names:
    new_name = name.strip().lower()
    cleaned_names.append(new_name)
```

```python
print("Cleaned Names:", cleaned_names)
```

```python
# Task 4 - Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]
```

```python
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)

    if length > 10:
        print("Long message detected")

    print("----------------")
```

```python
# Task 5 - Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
```

```python
error_count = 0
```

```python
for entry in logs:
    if entry == "ERROR":
        error_count += 1
```

```python
print("Total Error Messages:", error_count)
```
