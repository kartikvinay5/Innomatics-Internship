```python

# Problem 1: Smart Light Controller


class SmartLight:

    # Constructor to initialize light name
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status

    # Method to change light status
    def switch(self, action):
        if action.upper() == "ON":
            self.status = "ON"
        elif action.upper() == "OFF":
            self.status = "OFF"

    # Method to display current status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Sample Usage
light1 = SmartLight("Bedroom Light")
light1.switch("ON")
light1.display_status()
```

```python

# Problem 2: Employee ID Card System


class Employee:

    # Constructor to store employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Usage
emp1 = Employee("Rahul", "EMP102", "IT")
emp1.display_id_card()
```

```python

# Problem 3: Mobile Contact Record


class Contact:

    # Constructor to store contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact information
    def save_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Usage
contact1 = Contact("Anita", "9876543210")
contact1.save_contact()
```

```python
# Problem 4: Product Price Tag Generator

class Product:

    # Constructor to store product details
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    # Method to print formatted price tag
    def print_price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))


# Sample Usage
product1 = Product("Headphones", 2499)
product1.print_price_tag()
```

```python

# Problem 5: Movie Rating Display System


class Movie:

    # Constructor to store movie details
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    # Method to display rating
    def show_details(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


# Sample Usage
movie1 = Movie("Inception", 4.8)
movie1.show_details()
```

```python

# Problem 6: Delivery Address Manager


class Delivery:

    # Constructor to store customer details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def print_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Usage
delivery1 = Delivery("Suman", "Hyderabad")
delivery1.print_delivery_details()
```
